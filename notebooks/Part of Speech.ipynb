```python
import nltk

import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
df = pd.read_csv(
    '../singapore-streets.txt', 
    header=None, 
    names=['street name'],
)
df.head(3)
```

```python
def pos_tag_street_name(street_name):
    tokens = nltk.word_tokenize(street_name)
    pos_tags = nltk.pos_tag(tokens)
    return pos_tags

df['pos tags'] = df['street name'].apply(pos_tag_street_name)

df['words'] = df['pos tags'].apply(lambda x: [word for word, pos in x])
df['pos'] = df['pos tags'].apply(lambda x: [pos for word, pos in x])

df.head(10)
```

```python
pos_tag_mapping = {
    'CC': 'Coordinating conjunction',
    'CD': 'Cardinal number',
    'DT': 'Determiner',
    'EX': 'Existential there',
    'FW': 'Foreign word',
    'IN': 'Preposition or subordinating conjunction',
    'JJ': 'Adjective',
    'JJR': 'Adjective, comparative',
    'JJS': 'Adjective, superlative',
    'LS': 'List item marker',
    'MD': 'Modal',
    'NN': 'Noun, singular or mass',
    'NNS': 'Noun, plural',
    'NNP': 'Proper noun, singular',
    'NNPS': 'Proper noun, plural',
    'PDT': 'Predeterminer',
    'POS': 'Possessive ending',
    'PRP': 'Personal pronoun',
    'PRP$': 'Possessive pronoun',
    'RB': 'Adverb',
    'RBR': 'Adverb, comparative',
    'RBS': 'Adverb, superlative',
    'RP': 'Particle',
    'SYM': 'Symbol',
    'TO': 'to',
    'UH': 'Interjection',
    'VB': 'Verb, base form',
    'VBD': 'Verb, past tense',
    'VBG': 'Verb, gerund or present participle',
    'VBN': 'Verb, past participle',
    'VBP': 'Verb, non-3rd person singular present',
    'VBZ': 'Verb, 3rd person singular present',
    'WDT': 'Wh-determiner',
    'WP': 'Wh-pronoun',
    'WP$': 'Possessive wh-pronoun',
    'WRB': 'Wh-adverb'
}

def get_values(g):
    return [pos_tag_mapping[x] for x in g]

df['pos description'] = df['pos'].map(get_values)

df.head(10)
```

```python
df['pos'] = df['pos'].apply(lambda x: str(x))
df['pos description'] = df['pos description'].apply(lambda x: str(x))

df.groupby(['pos', 'pos description'])\
.size()\
.sort_values(ascending=False)\
.reset_index(name='Count')
```
